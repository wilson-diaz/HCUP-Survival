```python
file = '../datasets/FullHCUPDataset.csv'
out_file = 'HCUP-afib.csv'
newCSV = open(out_file, mode = 'a')

diabetes = 0
with open(file, mode='r') as csv_file:
    
    for line in csv_file:
        visits = line.replace('\n','').split('|')[2].split('#')
        for x in range(0, len(visits)):
            if "D_427" in visits[x]:
                diabetes+=1
                newCSV.write(line)
                break
                
newCSV.close()
print(diabetes) 
```

```python
codes = {}
index = 1
index_V = 5000
index_P = 10000
i = 1
out_file = 'HCUP-afib.csv'
with open(out_file) as diabetes_file:
    for patient in diabetes_file:
        visits = patient.replace('\n','').split('|')[2].split('#')
        for visit in visits:
            visit_split = visit.split(',')
            for v in visit_split:
                if v.startswith("P_"):
                    k = v[0 : 5]
                    if k not in codes:
                        codes[k] = index_P
                        index_P += 1
                        print(k, ' -- INDEX_P', index_P)
                if v.startswith("D_V"):
                    k = v[0: 6]
                    if k not in codes:
                        codes[k] = index_V
                        index_V += 1 
                        print(k, " -- INDEX_V", index_V)
                elif v.startswith("D_") and not v.startswith("D_V"):
                    k = v[0:5]
                    if k not in codes:
                        codes[k] = index
                        #print(d)
                        index += 1
                        print(k, " -- INDEX", index)
```

```python
import operator
codes_file = open('disease_codes.csv', 'w')
d_sorted = sorted(codes.items(), key=operator.itemgetter(1))
for i in d_sorted:
    codes_file.write(str(i) + "\n")
codes_file.close() 
```

```python
with open('disease_codes.csv', 'r') as f:
    for line in f:
        print(line)
```

```python
# get first 3 digits only. E.g., if D_250000 -> D_250
def get_general_code(med_code):
    if med_code.split('_')[1].startswith('V'):
        return med_code[0:6]
    return med_code[0:5]
```

```python
# index data and write a new file
new_csv = open('HCUP_indexed.csv', mode='w')
with open(out_file) as diabetes_file:
    for line in diabetes_file:
        front = line.replace('\n', '').split('|')[:2]
        visits = line.replace('\n','').split('|')[2].split('#')
        new_visits = []
        
        for visit in visits:
            link_and_age = visit.split(',')[:2]
            med_codes = visit.split(',')[2:-1]
            new_codes = []
            
            for med_code in med_codes:
                general_med_code = get_general_code(med_code)
                
                med_code = codes[general_med_code] # get index from dict
                
                new_codes.append(str(med_code))
            
            new_visits.append(','.join(link_and_age + new_codes))
        new_csv.write('|'.join(front) + '|' + '#'.join(new_visits) + '\n')
new_csv.close()
```

```python
# how many patients with more than 50 visits
# delete those patients who have more than 50 visits
num_patients_with_more_than_x_visits = 0

csv_excludes_more_than_50_visits = open('HCUP_excludes_more_than_50_visits_f.csv', mode='w')


num_visits = 0
data = 'HCUP_indexed.csv'
with open(data) as indexed_data:
    for row in indexed_data:
    
        visits = row.replace('\n','').split('|')[2].split('#')
        
        num_visits = 0
        for k in range(0,len(visits)):
            num_visits += 1
            
        if (num_visits > 50):
            num_patients_with_more_than_x_visits += 1
        else:
            csv_excludes_more_than_50_visits.write(row)
            
csv_excludes_more_than_50_visits.close()                
print(num_patients_with_more_than_x_visits)
```

```python
# remove procedure codes

data = 'HCUP_excludes_more_than_50_visits_f.csv'
csv_without_procedures = open('HCUP_indexed_ex_procedures.csv', mode='w')

with open(data) as diabetes_file:
    for line in diabetes_file:
        front = line.replace('\n', '').split('|')[:2]
        visits = line.replace('\n','').split('|')[2].split('#')
        new_visits = []
        
        for visit in visits:
            link_and_age = visit.split(',')[:2]
            med_codes = visit.split(',')[2:]
            #print(med_codes)
            #sys.exit()
            new_codes = []
            
            for med_code in med_codes:
                if int(med_code) >= 10000:
                    continue
                else:
                    new_codes.append(str(med_code))
            
            new_visits.append(','.join(link_and_age + new_codes))
        csv_without_procedures.write('|'.join(front) + '|' + '#'.join(new_visits) + '\n')
csv_without_procedures.close()
```
